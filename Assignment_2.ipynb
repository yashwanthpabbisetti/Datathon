# Histogram Graph for County vs Total Active Borrowers

```python
x=data['County']
y= data['Total Active Borrowers']
#giving the the columns x and y 
```

```python
plt.hist([x,y],bins=30,stacked=True,color=['red','yellow'],edgecolor='red')
plt.title('Total Active Borrowers vs County' )
plt.show()
#histogram
```

The graph compares the number of active borrowers across different counties. The x-axis represents the county (likely represented by a numeric code), while the y-axis shows the number of active borrowers. Most counties have relatively low numbers of borrowers, generally below 500, as indicated by the short blue and red bars spread across the graph. However, there is one striking outlier: the county represented at the far left (possibly county code 0 or 1) has an exceptionally high number of active borrowers, reaching over 3000, depicted by a tall yellow bar. This suggests a significant concentration of borrowing activity in one particular county, which dramatically outpaces all others shown. The stark contrast between this outlier and the rest of the counties implies there may be unique economic, demographic, or policy factors at play in that specific region driving much higher borrowing rates.

# Boxplot Graph for County vs Total Active Borrowers

```python
plt.boxplot([x, y])
plt.title('Box Plot of County vs Total Active Borrowers')
plt.show()
#Boxplot
```

The box figure contrasts two distinct categories—likely county groups or time periods—in terms of the distribution of Total Active Borrowers. With a box spanning from around 20,000 to 45,000 borrowers, the first category (1) displays a large range of numbers, indicating notable variability. The line in the box represents its median, which is approximately 30,000. The whiskers show that some counties have relatively few borrowers while others have numerous, ranging from about 0 to approximately 80,000. The distribution of the second group (2), on the other hand, is substantially more compact, with all data points closely clustered around 1,000 borrowers. There are multiple outliers in this category that are shown as individual points and are above the upper whisker.
